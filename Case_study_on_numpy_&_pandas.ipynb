```python
import numpy as np
import pandas as pd
```

##Q1. Employee salary adjustment

```python
salaries = np.array([45000, 52000, 61000, 48000, 75000])
new_salaries = salaries * 1.10
print('1. New salaries: ',new_salaries)

average_salary = salaries.mean()
print('2. Average salary: ',average_salary)

min_salary = salaries.min()
print('3.a. Minimum salary: ', min_salary)

max_salary = salaries.max()
print('3.b. Minimum salary: ', max_salary)

print('4. Salaries greater that 60000: ', salaries[salaries > 60000])
```

##Q2. Sales Data Analysis

```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])

print('1. Total sales: ',sales.sum());

print("2.a. Mean: ", sales.mean())
print('2.b. Standard deviation: ',np.std(sales))
print (pd.Series(sales).std())

min_sales = np.min(sales)
max_sales = np.max(sales)

# Min-Max scaling
normalized_sales = (sales - min_sales) / (max_sales - min_sales)

print('4. Normalized sales data: ',normalized_sales)
```

##Q3. Matrix Operations for Business Data

```python
sales_matrix = np.array([
    [120, 140, 130, 150],
    [80, 90, 100, 95],
    [200, 210, 190, 205]
])

sales_per_product = sales_matrix.sum(axis=1)
print(f"1. Total sales per product: {sales_per_product}")

sales_per_region = sales_matrix.sum(axis=0)
print(f"2. Total sales per region: {sales_per_region}")

best_product_idx = np.argmax(sales_per_product)
print(f"3. Best performing product is at row index {best_product_idx} with {sales_per_product[best_product_idx]} sales.")

best_region_idx = np.argmax(sales_per_region)
print(f"4. Region with highest sales is at column index {best_region_idx} with {sales_per_region[best_region_idx]} sales.")
```

##Q4. Employee Dataset Exploration

```python
data = {
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'Finance'],
    'Salary': [50000, 65000, 70000, 72000, 52000, 68000],
    'Experience': [2, 5, 7, 6, 3, 4]
}
df = pd.DataFrame(data)

#2. Basic statistics
print('2. Basic statistics: \n',df.describe(),'\n')

#3. Average salary per department
print('3. Average Salary Per Department.\n',df.groupby('Department')['Salary'].mean())

#4. Employees with experience greater than 4 years
print('\n4. Employees with Experience > 4 Years.\n',df[df['Experience'] > 4])

# 5. Add a new column Bonus = 10% of salary
df['Bonus'] = df['Salary'] * 0.10
print("\n5. New DataFrame with new column for bonus.")
print(df)
```

##Q5. Missing Data Handling

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda', 'David'],
    'Age': [28, np.nan, 35, 32, np.nan],
    'Salary': [50000, 54000, np.nan, 58000, 60000]
}
df = pd.DataFrame(data)

# 1. Missing values
print('1. Missing Values Count\n',df.isna().sum())

# 2. Replace missing Age with mean age
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

# 3. Replace missing Salary with median salary
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

# 4. Remove rows with any remaining missing values
df.dropna(inplace=True)

print("\nFinal Cleaned DataFrame")
print(df)
```

##Q6. Sales Performance Analysis

```python
sales_data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Salesperson': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Sales': [12000, 15000, 13000, 17000, 16000, 14000, 15500, 16500]
}
df = pd.DataFrame(sales_data)

# 1. Find total sales per region
print("1. Total Sales per Region")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False))

# 2. Identify the top salesperson
top_person = df.loc[df['Sales'].idxmax()]
print("\n2. Top Salesperson")
print(f"Salesperson {top_person['Salesperson']} with {top_person['Sales']} sales.")

# 3. Compute average sales across regions
print("\n3. Average Sales")
print(f"Overall Average: {df['Sales'].mean()}")
print("\nAverage Per Region:")
print(df.groupby('Region')['Sales'].mean().sort_values(ascending=False))

# 4. Sort the dataset by sales descending
sorted_df = df.sort_values(by='Sales', ascending=False)
print("\n4. Sorted Dataset")
print(sorted_df)
```

##Q7. Customer Purchase Analysis

```python
data = {
    'CustomerID': [101, 102, 103, 104, 105, 106, 107],
    'Age': [25, 45, 35, 50, 23, 40, 60],
    'PurchaseAmount': [250, 600, 400, 800, 150, 700, 900],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi', 'Chennai']
}
df = pd.DataFrame(data)

# 1. Average purchase amount by city
print("1. Average Purchase by City")
print(df.groupby('City')['PurchaseAmount'].mean().sort_values(ascending=False))

# 2. Customers with purchase above overall average
overall_avg = df['PurchaseAmount'].mean()
high_spenders = df[df['PurchaseAmount'] > overall_avg]
print(f"\n2. High Spenders (Above {overall_avg:.2f})")
print(high_spenders)

# 3. City generating highest revenue
revenue_by_city = df.groupby('City')['PurchaseAmount'].sum()
top_city = revenue_by_city.idxmax()
print(f"\n3. City with Highest Revenue")
print(f"{top_city} with {revenue_by_city[top_city]} total sales")

# 4. Create a column Age_Group
conditions = [
    (df['Age'] < 30),
    (df['Age'] >= 30) & (df['Age'] <= 50),
    (df['Age'] > 50)
]
choices = ['Young', 'Adult', 'Senior']
df['Age_Group'] = np.select(conditions, choices, default='Unknown')

print("\n4. Final Dataset with Age Groups")
print(df)

# 5. Average purchase per age group
print("\n5. Average Purchase by Age Group")
print(df.groupby('Age_Group')['PurchaseAmount'].mean())
```

##Q8. Time Series Sales Analysis

```python
dates = pd.date_range('2024-01-01', periods=12, freq='ME')
sales = [200, 220, 250, 270, 300, 320, 350, 360, 380, 400, 420, 450]
df = pd.DataFrame({'Date': dates, 'Sales': sales})

# 1. Convert Date to datetime index
df.set_index('Date', inplace=True)

# 2. Compute monthly percentage growth
df['Monthly_Growth_Pct'] = df['Sales'].pct_change() * 100
print('\n2. Monthly percentage growth\n',df['Monthly_Growth_Pct'])

# 3. Calculate 3-month rolling average sales
df['3_Month_Rolling_Avg'] = df['Sales'].rolling(window=3).mean()
print('\n3. 3 month rolling average\n',df['3_Month_Rolling_Avg'])

# 4. Identify the month with highest growth rate
highest_growth_month = df['Monthly_Growth_Pct'].idxmax()
highest_growth_value = df['Monthly_Growth_Pct'].max()

print("\n4. Month with Highest Growth Rate")
print(f"{highest_growth_month.strftime('%B %Y')} with {highest_growth_value:.2f}% growth")

print("\nFinal DataFrame")
print(df.round(2)) # Rounding to 2 decimal places for cleaner output
```
